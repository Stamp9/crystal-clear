## Visualize contracts stats

This notebook collects the data from `data` directory and plots graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

Visualize the number of alive contracts by year

```python

alive_csv = '../data/alive_contracts_number.csv'
data = pd.read_csv(alive_csv)

plt.figure(figsize=(10, 7))
bars = plt.bar(data['year'], data['n_alive_contracts'], color='skyblue')

# Add labels and title
plt.xlabel('Year', fontsize=16, fontweight='bold')
plt.ylabel('Number of Alive Contracts', fontsize=16, fontweight='bold')
plt.title('Number of Alive Contracts by Year', fontsize=18, fontweight='bold')

# Add labels on top of each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of each bar (the number of contracts)
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')  # Place the label

# Set x-ticks to show every year with larger text
plt.xticks(data['year'], rotation=45, fontsize=14, fontweight='bold')

# Set y-ticks with larger text
plt.yticks(fontsize=14, fontweight='bold')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x/1e6)}M'))

# Display the plot
plt.savefig("figures/alive_contracts_by_year.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
