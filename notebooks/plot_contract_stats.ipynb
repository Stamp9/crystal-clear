## Visualize contracts stats

This notebook collects the data from `data` directory and plots graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

Visualize the number of alive contracts by year

```python

alive_csv = '../data/raw/alive_contracts_year.csv'
data = pd.read_csv(alive_csv)

plt.figure(figsize=(10, 7))
bars = plt.bar(data['year'], data['n_alive_contracts'], color='skyblue')

# Add labels and title
plt.xlabel('Year', fontsize=16, fontweight='bold')
plt.ylabel('Number of Alive Contracts', fontsize=16, fontweight='bold')
plt.title('Number of Alive Contracts by Year', fontsize=18, fontweight='bold')

# Add labels on top of each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of each bar (the number of contracts)
    # plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')  # Place the label
    plt.text(
        bar.get_x() + bar.get_width() / 2, 
        yval, 
        f'{yval / 1e6:.3f}M',  # Convert to millions and format to one decimal place
        ha='center', 
        va='bottom', 
        fontsize=12, 
        fontweight='bold'
    )

# Set x-ticks to show every year with larger text
plt.xticks(data['year'], rotation=45, fontsize=14, fontweight='bold')

# Set y-ticks with larger text
plt.yticks(fontsize=14, fontweight='bold')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x/1e6)}M'))

# Display the plot
plt.savefig("../data/figures/alive_contracts_by_year.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
# Load data
alive_csv = '../data/raw/alive_contracts_year.csv'
data = pd.read_csv(alive_csv)

# Create figure
plt.figure(figsize=(10, 7))

# Plot area chart
plt.fill_between(data['year'], data['n_alive_contracts'], color='skyblue', alpha=0.6)

# Add labels and title
plt.xlabel('Year', fontsize=16, fontweight='bold')
plt.ylabel('Number of Alive Contracts', fontsize=16, fontweight='bold')
plt.title('Number of Alive Contracts by Year', fontsize=18, fontweight='bold')

# Add labels on top of the area plot
for x, y in zip(data['year'], data['n_alive_contracts']):
    plt.text(x, y, f'{y:,}', ha='center', va='bottom', fontsize=12, fontweight='bold')

# Set x-ticks to show every year with larger text
plt.xticks(data['year'], rotation=45, fontsize=14, fontweight='bold')

# Set y-ticks with larger text
plt.yticks(fontsize=14, fontweight='bold')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x/1e6)}M'))

# Save and display the plot
plt.savefig("../data/figures/alive_contracts_by_year.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
calls_tx_csv = '../data/raw/contracts_per_tx_year.csv'
data = pd.read_csv(calls_tx_csv)

# Convert to DataFrame
df = pd.DataFrame(data)

# Update plot parameters
plt.rcParams.update({
    'font.size': 18,  # Increased base font size
    'axes.labelsize': 18,  # Larger axis labels
    'axes.labelweight': 'bold',
    'axes.titlesize': 18,  # Larger title
    'xtick.labelsize': 16,  # Larger tick labels
    'ytick.labelsize': 16,
    'legend.fontsize': 16,
    'font.weight': 'bold'  # Make all text bold
})

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot 'Average' values on the primary y-axis
ax1.plot(df["year"], df["avg"], marker="o", linestyle="-", color="blue", label="Average")
ax1.set_xlabel("Year")
ax1.set_ylabel("Avg. contracts per tx", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# # Add value labels for 'Average'
# for i, txt in enumerate(df["avg"]):
#     ax1.text(df["year"][i], txt, f"{txt:.2f}", ha='right', va='bottom', fontsize=9, color="blue")

# Create a secondary y-axis for 'Max' values
ax2 = ax1.twinx()
ax2.plot(df["year"], df["max"], marker="o", linestyle="-", color="red", label="Max")
ax2.set_ylabel("Max. contracts per tx", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# # Add value labels for 'Max'
# for i, txt in enumerate(df["max"]):
#     ax2.text(df["year"][i], txt, f"{txt}", ha='left', va='bottom', fontsize=9, color="red")

# Title and layout
plt.title("Average and Maximum contracts called per transaction by year", fontsize=16, fontweight='bold')
plt.grid(True, linestyle="--", alpha=0.5)  # Add light grid for readability

plt.savefig("../data/figures/contracts_per_tx_year.pdf", format='pdf', bbox_inches='tight')

# Show plot
plt.show()
```
